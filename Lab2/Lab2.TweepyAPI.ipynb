We have defined our keys in a external file named "TwitterKeys.json" file, so first we need to read it and parse it:

```python
import json
jsonFileText = None
with open("TwitterKeys.json", 'r') as inputFile: jsonFileText = inputFile.read().replace('\n', '')
keys = json.loads(jsonFileText)
```

Now we can make use of them:

```python
import tweepy
from tweepy import OAuthHandler

consumer_key = keys["consumer_key"]
consumer_secret = keys["consumer_secret"]
access_token = keys["access_token"]
access_secret = keys["access_secret"]

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
 
api = tweepy.API(auth)
```

```python
user = api.me()
 
print('Name: ' + user.name)
print('Location: ' + user.location)
print('Friends: ' + str(user.followers_count))
print('Created: ' + str(user.created_at))
print('Description: ' + str(user.description))
```

```python
for tweet in tweepy.Cursor(api.user_timeline).items(1):
    for key, value in tweet._json.iteritems():
        if isinstance(tweet._json[key], dict):
            print key
            for key2, value2 in tweet._json[key].iteritems():
                print "\t", key2 
        else:
            print key
```

Now we could look for the tweets themselves and their creation date:

```python
for tweet in tweepy.Cursor(api.user_timeline).items():
    tweetJSON = tweet._json
    print tweetJSON["user"]["name"], "at", tweetJSON["created_at"], ":", tweetJSON["text"]
```
